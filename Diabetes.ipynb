````python
import pandas as pd
````

````python
df = pd.read_csv('/content/diabetes.csv')
df
````

````python
X = df.drop(columns=['Outcome'])
X
````

````python
Y = df['Outcome']
Y
````

````python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=42)

knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
````

````python
y_predict=knn.predict(xtest)
y_predict
````

````python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(ytest,y_predict)
print("Confusion matrix \n", confusion_matrix)
````

````python
tn=confusion_matrix[0][0]
fp=confusion_matrix[0][1]
fn=confusion_matrix[1][0]
tp=confusion_matrix[1][1]
````

````python
accuracy = (tp+tn)/(tn+fp+fn+tp)
error = (fp+fn)/(tn+fp+fn+tp)
precision= (tp)/(tp+fp)
recall = (tp)/ (tp+fn)
print("\nBy using manual calculation : ")
print("Accuracy : ", accuracy)
print("Error rate : ", error)
print("Precision : ", precision)
print("Recall : ",recall)

print("\nBy using inbuilt function : ")
print("Accuracy : ", metrics.accuracy_score(ytest,y_predict))
print("Error rate : ", 1-accuracy)
print("Precision : ", metrics.precision_score(ytest,y_predict))
print("Recall : ",metrics.recall_score(ytest,y_predict))
````
