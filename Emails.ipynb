````python
import pandas as pd
````

````python
df = pd.read_csv('/content/emails.csv')
df
````

````python
x= df.iloc[:, 1:3001]
x
````

````python
y= df.iloc[:,-1]
y
````

````python
y_knn_predict = knn_model.predict(x_test)
y_knn_predict
````

# Support Vector Machine (SVM)

## Overview:
Support Vector Machine (SVM) is a powerful supervised machine learning algorithm primarily used for classification tasks, although it can also be used for regression. SVM works by finding the optimal hyperplane that separates the data points of different classes in a high-dimensional feature space.

## Key Concepts:
- **Hyperplane**: A hyperplane is a decision boundary that separates data points of different classes. For binary classification, the hyperplane is a line (in 2D), a plane (in 3D), or a higher-dimensional hyperplane in higher dimensions.
- **Support Vectors**: Support vectors are the data points that are closest to the hyperplane. These data points influence the position and orientation of the hyperplane.
- **Margin**: The margin is the distance between the hyperplane and the support vectors. SVM aims to maximize this margin to create the most robust classifier.

## How SVM Works:
1. **Linear SVM**: If the data is linearly separable, SVM will find a linear hyperplane that maximizes the margin between the two classes.
2. **Non-linear SVM**: When the data is not linearly separable, SVM uses the **kernel trick** to transform the data into a higher-dimensional space where it becomes separable. Common kernel functions include polynomial, radial basis function (RBF), and sigmoid kernels.

## Example of SVM:

Let's take an example where we have two classes of data: `Class 1` and `Class 2`. We want to classify these classes using a linear SVM. Here's how it can be done:

```python
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([[2, 3], [3, 3], [4, 4], [5, 6], [5, 8], [7, 8], [8, 9], [6, 4]])  # Features
y = [0, 0, 0, 1, 1, 1, 1, 0]  # Labels (0: Class 1, 1: Class 2)

# Creating the SVM model
model = SVC(kernel='linear')
model.fit(X, y)

# Plotting the decision boundary
xx, yy = np.meshgrid(np.linspace(0, 10, 100), np.linspace(0, 10, 100))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.75)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', s=100, cmap=plt.cm.Paired)
plt.title("SVM Classifier with Linear Kernel")
plt.show()


K-Nearest Neighbors (KNN)
Overview:
K-Nearest Neighbors (KNN) is a simple and intuitive supervised machine learning algorithm used for classification and regression tasks. It works by finding the K nearest data points to a query point and making predictions based on the majority class (for classification) or the average value (for regression) of those neighbors.

Key Concepts:
Distance Metric: KNN uses a distance metric to measure the similarity between data points. Common distance metrics include Euclidean, Manhattan, and Minkowski distance.
K (Number of Neighbors): The value of K represents the number of neighbors to consider when making a prediction. A smaller value of K may lead to overfitting, while a larger value may smooth out the decision boundary, possibly underfitting the model.
How KNN Works:
For Classification: When a new data point (query point) is to be classified, KNN looks at the K nearest neighbors and assigns the query point to the class that is most common among these neighbors.
For Regression: In regression tasks, KNN takes the average of the K nearest neighbors' values to predict the value for the query point.
Example of KNN:
Let’s say we have a dataset of points belonging to two different classes, and we want to classify a new point based on its nearest neighbors:

Copy code
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([[2, 3], [3, 3], [4, 4], [5, 6], [5, 8], [7, 8], [8, 9], [6, 4]])  # Features
y = [0, 0, 0, 1, 1, 1, 1, 0]  # Labels (0: Class 1, 1: Class 2)

# Creating the KNN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# Predicting for a new point
new_point = np.array([[6, 7]])
prediction = knn.predict(new_point)

# Plotting the data and the prediction
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', s=100, cmap=plt.cm.Paired)
plt.scatter(new_point[0][0], new_point[0][1], c='red', marker='x', s=200, label='New Point')
plt.title(f"KNN Prediction: {prediction[0]}")
plt.legend()
plt.show()
In this example:

The KNeighborsClassifier class from sklearn.neighbors is used to create the KNN model.
The model is trained using the fit() function, and n_neighbors=3 specifies that the algorithm should look at the 3 nearest neighbors to make a prediction.
We predict the class of a new point (red "x" mark) based on its 3 nearest neighbors.
Conclusion:
SVM is highly effective when the data is linearly separable or when you can transform data into a higher-dimensional space using kernels. It is particularly good at handling high-dimensional spaces and outliers.
KNN, on the other hand, is simple and intuitive. It doesn't require a training phase but can be computationally expensive during the prediction phase, especially with large datasets. It performs well when the decision boundary is irregular and non-linear.
